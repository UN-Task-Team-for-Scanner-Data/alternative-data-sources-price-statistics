```python
import pandas as pd

# Load data
df = pd.read_csv("monthly_avg_price.csv")  # change name if needed
df.head()
```

```python
import numpy as np

np.random.seed(42)  # for reproducibility

# Let's assume your price column is called 'price'
# We'll create quantities inversely related to price
max_price = df['price'].max()
min_price = df['price'].min()

# Scale prices to 0â€“1 range
price_scaled = (df['price'] - min_price) / (max_price - min_price)

# Higher price = lower quantity, add some randomness
df['quantity'] = (1 - price_scaled) * np.random.randint(20, 800, size=len(df))

# Ensure no zero quantities
df['quantity'] = df['quantity'].clip(lower=1).astype(int)
df.head()
```

```python
df.to_csv("monthly_with_qty.csv", index=False)
```
